```python
import skimage
import matplotlib.pyplot as plt
import skimage.io
```

```python
I = skimage.data.camera()
ROI = I[100:200, 200:300]
plt.imshow(ROI, 'gray')
plt.show()
```

```python
r = I[::10, ::10]
print(r.shape)
plt.imshow(r, 'gray')
plt.show()
```

```python
for s in [1, 2, 3, 4, 5, 10, 20, 50]:
    r = I[::s, ::s]
    print(r.shape)
    plt.imshow(r, 'gray')
    plt.show()
```

```python
for s in [1, 2, 3, 4, 5, 10, 20, 50]:
    r = I[::s, ::s]
    skimage.io.imsave("test{}.png".format(s), r)
```

```python
for s in [1, 2, 3, 4, 5, 10, 20, 50]:
    plt.figure(figsize=(8/s, 8/s))
    plt.imshow(I, 'gray')
    plt.show()
```

```python
for s in [1, 2, 3, 4, 5, 10, 20, 50]:
    
    plt.figure(figsize=(8/s, 8/s))
    r = I[::s, ::s]
    plt.imshow(r, 'gray')
    plt.show()
```
